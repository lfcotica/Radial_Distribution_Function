# **What are Radial Distribution Functions?**
The radial distribution function (RDF) (or pair correlation function) characterises the structure of a system of particles. If we select an arbitrary particle as the origin, the RDF describes the number of particles we would observe relative to the bulk density of the system, as a function of distance. This is calculated and averaged over every particle in the structure being considered.

https://en.wikipedia.org/wiki/Radial_distribution_function
The radial distribution function is usually determined by calculating the distance between all particle pairs and binning them into a histogram.

```python
import numpy as np
from rdfpy import rdf
from pymatgen.core import Structure
```

cif Ti - https://legacy.materialsproject.org/materials/mp-46/

We then load the structure and create a supercell so that there are enough atoms in the crystal to compute an RDF from. Also, to make the resulting function smoother, we add some noise to the coordinates.

```python
structure = Structure.from_file('/content/drive/MyDrive/database/Ti_mp-46_computed.cif')
structure.make_supercell(10)

#coords = structure.cart_coords
#noise = np.random.normal(loc=0.0, scale=0.05, size=(coords.shape))
#coords = coords + noise
```

```python
coords = structure.cart_coords
```

```python
coords
```

```python
noise = np.random.normal(loc=0.0, scale=0.05, size=(coords.shape))
coords = coords + noise
```

```python
g_r, radii = rdf(coords, dr=0.05)
```

```python
import matplotlib.pyplot as plt
X = radii
Y = g_r
plt.plot(X, Y, lw=2)

# naming the x axis
plt.xlabel('r')
# naming the y axis
plt.ylabel('g(r)')

# giving a title to my graph
plt.title('Radial Distribution Function (RDF) for Ti')

# function to show the plot
plt.show()
```

# **BiFeO3 R3c** - https://legacy.materialsproject.org/materials/mp-23501/#

```python
structure_r3c = Structure.from_file('/content/drive/MyDrive/database/FeBiO3_mp-23501_computed.cif')
structure_r3c.make_supercell(10)

coords_r3c = structure_r3c.cart_coords
noise_r3c = np.random.normal(loc=0.0, scale=0.05, size=(coords_r3c.shape))
coords_r3c = coords_r3c + noise_r3c
```

```python
g_r_r3c, radii_r3c = rdf(coords_r3c, dr=0.05)
```

```python
import matplotlib.pyplot as plt
X_r3c = radii_r3c
Y_r3c = g_r_r3c
plt.plot(X_r3c, Y_r3c, lw=2, label = 'R3c')

# naming the x axis
plt.xlabel('r')
# naming the y axis
plt.ylabel('g(r)')
plt.xlim([0, 10])

# giving a title to my graph
plt.title('Radial Distribution Function (RDF) for BiFeO3  - R3c')

# function to show the plot
plt.show()
```
